#### 2. Identify the top 5 states with the highest penetration rate in 2-wheeler and 4-wheeler EV sales in FY 2024.
Basically the question means Top 5 States with Highest Penetration Rate in 2-Wheeler and 4-Wheeler EV Sales (FY 2024)

```python
import pandas as pd

# Load the data
data = pd.read_csv('electric_vehicle_sales_by_state.csv')

# Convert the date column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Define the fiscal year range for FY 2024
fy_2024_start = '2023-04-01'
fy_2024_end = '2024-03-31'

# Filter the data for FY 2024
data_fy_2024 = data[(data['date'] >= fy_2024_start) & (data['date'] <= fy_2024_end)]

# Calculate penetration rates for 2-wheelers
penetration_2W = data_fy_2024[data_fy_2024['vehicle_category'] == '2-Wheelers']
penetration_2W = penetration_2W.groupby('state').agg({
    'electric_vehicles_sold': 'sum',
    'total_vehicles_sold': 'sum'
}).reset_index()

penetration_2W['penetration_rate_2W'] = (penetration_2W['electric_vehicles_sold'] / penetration_2W['total_vehicles_sold']) * 100

# Identify top 5 states with the highest penetration rate for 2-wheelers
top_5_states_2W = penetration_2W.sort_values(by='penetration_rate_2W', ascending=False).head(5)

# Calculate penetration rates for 4-wheelers
penetration_4W = data_fy_2024[data_fy_2024['vehicle_category'] == '4-Wheelers']
penetration_4W = penetration_4W.groupby('state').agg({
    'electric_vehicles_sold': 'sum',
    'total_vehicles_sold': 'sum'
}).reset_index()

penetration_4W['penetration_rate_4W'] = (penetration_4W['electric_vehicles_sold'] / penetration_4W['total_vehicles_sold']) * 100

# Identify top 5 states with the highest penetration rate for 4-wheelers
top_5_states_4W = penetration_4W.sort_values(by='penetration_rate_4W', ascending=False).head(5)

# Display results
print("Top 5 States with Highest Penetration Rate in 2-Wheeler EV Sales (FY 2024):")
print(top_5_states_2W)

print("\nTop 5 States with Highest Penetration Rate in 4-Wheeler EV Sales (FY 2024):")
print(top_5_states_4W)

# Set up the plot
plt.figure(figsize=(14, 8))

# Top 5 States for 2-Wheeler Penetration Rates
plt.subplot(1, 2, 1)
sns.barplot(x='penetration_rate_2W', y='state', data=top_5_states_2W, palette='coolwarm')
plt.title('Top 5 States with Highest Penetration Rate in 2-Wheeler EV Sales (FY 2024)')
plt.xlabel('Penetration Rate (%)')
plt.ylabel('State')

# Top 5 States for 4-Wheeler Penetration Rates
plt.subplot(1, 2, 2)
sns.barplot(x='penetration_rate_4W', y='state', data=top_5_states_4W, palette='coolwarm')
plt.title('Top 5 States with Highest Penetration Rate in 4-Wheeler EV Sales (FY 2024)')
plt.xlabel('Penetration Rate (%)')
plt.ylabel('State')

plt.tight_layout()
plt.show()
```

#### 3. List the states with negative penetration (decline) in EV sales from 2022 to 2024?
Basically the question means States with Declining EV Sales (2022 to 2024)

```python
import pandas as pd

# Load the data
data = pd.read_csv('electric_vehicle_sales_by_state.csv')

# Convert the date column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Define fiscal year ranges
fy_2022_start = '2022-04-01'
fy_2022_end = '2023-03-31'
fy_2024_start = '2023-04-01'
fy_2024_end = '2024-03-31'

# Filter data for FY 2022 and FY 2024
data_2022 = data[(data['date'] >= fy_2022_start) & (data['date'] <= fy_2022_end)]
data_2024 = data[(data['date'] >= fy_2024_start) & (data['date'] <= fy_2024_end)]

# Calculate penetration rates for FY 2022
penetration_2022 = data_2022.groupby(['state', 'vehicle_category']).agg({
    'electric_vehicles_sold': 'sum',
    'total_vehicles_sold': 'sum'
}).reset_index()
penetration_2022['penetration_rate_2022'] = penetration_2022['electric_vehicles_sold'] / penetration_2022['total_vehicles_sold']

# Calculate penetration rates for FY 2024
penetration_2024 = data_2024.groupby(['state', 'vehicle_category']).agg({
    'electric_vehicles_sold': 'sum',
    'total_vehicles_sold': 'sum'
}).reset_index()
penetration_2024['penetration_rate_2024'] = penetration_2024['electric_vehicles_sold'] / penetration_2024['total_vehicles_sold']

# Merge penetration rates for FY 2022 and FY 2024
merged_penetration = pd.merge(penetration_2022, penetration_2024, on=['state', 'vehicle_category'], suffixes=('_2022', '_2024'))

# Calculate the change in penetration rates
merged_penetration['penetration_rate_change'] = merged_penetration['penetration_rate_2024'] - merged_penetration['penetration_rate_2022']

# Identify states with declining penetration rates
declining_states = merged_penetration[merged_penetration['penetration_rate_change'] < 0]

# Display results
print("States with Declining Penetration Rates (2022 to 2024):")
print(declining_states[['state', 'vehicle_category', 'penetration_rate_2022', 'penetration_rate_2024', 'penetration_rate_change']])

plt.figure(figsize=(14, 8))

# Bar plot for declining penetration rates
sns.barplot(x='penetration_rate_change', y='state', hue='vehicle_category', data=declining_states, palette='coolwarm')
plt.title('States with Declining Penetration Rates in EV Sales (2022 to 2024)')
plt.xlabel('Change in Penetration Rate')
plt.ylabel('State')
plt.axvline(0, color='k', linestyle='--')

plt.show()
```

The questions means Quarterly Trends for Top 5 EV Makers (4-Wheelers) from 2022 to 2024

```python
# Load the data
data = pd.read_csv('electric_vehicle_sales_by_makers.csv')

# Convert the date column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Filter the data for 4-wheelers and the specified fiscal years
data_4w = data[(data['vehicle_category'] == '4-Wheelers') & 
               (data['date'] >= '2022-01-01') & 
               (data['date'] <= '2024-12-31')]

# Extract the fiscal year and quarter
data_4w['quarter'] = data_4w['date'].dt.to_period('Q')
data_4w['fiscal_year'] = data_4w['date'].dt.to_period('A-JUN')

# Identify the top 5 makers based on total sales volume for 4-wheelers from 2022 to 2024
top_makers = data_4w.groupby('maker')['electric_vehicles_sold'].sum().reset_index()
top_5_makers = top_makers.sort_values(by='electric_vehicles_sold', ascending=False).head(5)

# Filter data for top 5 makers
data_top_5 = data_4w[data_4w['maker'].isin(top_5_makers['maker'])]

# Aggregate sales data by quarter
quarterly_sales = data_top_5.groupby(['maker', 'quarter'])['electric_vehicles_sold'].sum().reset_index()

# Convert quarter to string for plotting
quarterly_sales['quarter'] = quarterly_sales['quarter'].astype(str)

# Check the quarterly sales data
print(quarterly_sales.head())

# Plotting the trends using Seaborn
plt.figure(figsize=(14, 8))
sns.lineplot(data=quarterly_sales, x='quarter', y='electric_vehicles_sold', hue='maker', marker='o')
plt.title('Quarterly Sales Trends for Top 5 EV Makers (4-Wheelers) from 2022 to 2024')
plt.xlabel('Quarter')
plt.ylabel('Electric Vehicles Sold')
plt.legend(title='Maker')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 5. How do the EV sales and penetration rates in Delhi compare to Karnataka for 2024?
It means Comparison of EV Sales and Penetration Rates: Delhi vs Karnataka (2024)


```python
# Load the data
data = pd.read_csv('electric_vehicle_sales_by_state.csv')

# Convert the date column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Filter data for 2024
data_2024 = data[data['date'].dt.year == 2024]

# Extract data for Delhi and Karnataka
data_delhi_2024 = data_2024[data_2024['state'] == 'Delhi']
data_karnataka_2024 = data_2024[data_2024['state'] == 'Karnataka']

# Calculate penetration rates for Delhi
penetration_rate_delhi = data_delhi_2024.groupby('vehicle_category').apply(
    lambda x: x['electric_vehicles_sold'].sum() / x['total_vehicles_sold'].sum()
).reset_index(name='penetration_rate')

# Calculate penetration rates for Karnataka
penetration_rate_karnataka = data_karnataka_2024.groupby('vehicle_category').apply(
    lambda x: x['electric_vehicles_sold'].sum() / x['total_vehicles_sold'].sum()
).reset_index(name='penetration_rate')

# Calculate total EV sales for Delhi
total_sales_delhi = data_delhi_2024.groupby('vehicle_category')['electric_vehicles_sold'].sum().reset_index()

# Calculate total EV sales for Karnataka
total_sales_karnataka = data_karnataka_2024.groupby('vehicle_category')['electric_vehicles_sold'].sum().reset_index()

# Merge penetration rates with total sales
comparison_delhi = pd.merge(total_sales_delhi, penetration_rate_delhi, on='vehicle_category')
comparison_karnataka = pd.merge(total_sales_karnataka, penetration_rate_karnataka, on='vehicle_category')

# Display the data
print("Comparison Data for Delhi (2024):")
print(comparison_delhi)
print("\nComparison Data for Karnataka (2024):")
print(comparison_karnataka)

# Plotting the results
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot total EV sales
sns.barplot(x='vehicle_category', y='electric_vehicles_sold', data=comparison_delhi, ax=axes[0], color='blue', alpha=0.6, label='Delhi')
sns.barplot(x='vehicle_category', y='electric_vehicles_sold', data=comparison_karnataka, ax=axes[0], color='orange', alpha=0.6, label='Karnataka')
axes[0].set_title('Total EV Sales in 2024')
axes[0].set_xlabel('Vehicle Category')
axes[0].set_ylabel('Electric Vehicles Sold')
axes[0].legend()

# Plot penetration rates
sns.barplot(x='vehicle_category', y='penetration_rate', data=comparison_delhi, ax=axes[1], color='blue', alpha=0.6, label='Delhi')
sns.barplot(x='vehicle_category', y='penetration_rate', data=comparison_karnataka, ax=axes[1], color='orange', alpha=0.6, label='Karnataka')
axes[1].set_title('EV Penetration Rates in 2024')
axes[1].set_xlabel('Vehicle Category')
axes[1].set_ylabel('Penetration Rate')
axes[1].legend()

plt.tight_layout()
plt.show()
```

#### 6. List down the compounded annual growth rate (CAGR) in 4-wheeler units for the top 5 makers from 2022 to 2024.
CAGR for Top 5 Makers (4-Wheelers, 2022-2024)
Calculate the CAGR for the top 5 4-wheeler makers using the formula:
CAGR=( EV start EV end) n11
(Ending Value/Beginning Value) ^ (1/No. of Periods)  1.

```python
# Load the data
data = pd.read_csv('electric_vehicle_sales_by_makers.csv')

# Convert the date column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Filter the data for 4-wheelers
data_4w = data[data['vehicle_category'] == '4-Wheelers']

# Filter data for 2022 and 2024
data_2022 = data_4w[data_4w['date'].dt.year == 2022]
data_2024 = data_4w[data_4w['date'].dt.year == 2024]

# Group by maker and sum the electric vehicles sold for 2022 and 2024
sales_2022 = data_2022.groupby('maker')['electric_vehicles_sold'].sum().reset_index()
sales_2024 = data_2024.groupby('maker')['electric_vehicles_sold'].sum().reset_index()

# Merge the sales data for 2022 and 2024
merged_sales = pd.merge(sales_2022, sales_2024, on='maker', suffixes=('_2022', '_2024'))

# Determine the top 5 makers in 2022
top_5_makers = merged_sales.nlargest(5, 'electric_vehicles_sold_2022')

# Calculate CAGR for the top 5 makers
def calculate_cagr(start, end, periods):
    return (end / start) ** (1 / periods) - 1

top_5_makers['CAGR'] = top_5_makers.apply(lambda row: calculate_cagr(row['electric_vehicles_sold_2022'], 
                                                                     row['electric_vehicles_sold_2024'], 
                                                                     2), axis=1)

# Display the results
print("CAGR for Top 5 4-Wheeler Makers (2022-2024):")
print(top_5_makers[['maker', 'electric_vehicles_sold_2022', 'electric_vehicles_sold_2024', 'CAGR']])

# Plotting the CAGR
plt.figure(figsize=(10, 6))
plt.bar(top_5_makers['maker'], top_5_makers['CAGR'], color='skyblue')
plt.xlabel('Maker')
plt.ylabel('CAGR (2022-2024)')
plt.title('CAGR of Top 5 4-Wheeler EV Makers (2022-2024)')
plt.xticks(rotation=45)
plt.show()
```

#### 7. List down the top 10 states that had the highest compounded annual growth rate (CAGR) from 2022 to 2024 in total vehicles sold.
Which means Top 10 States with Highest CAGR in Total Vehicles Sold (2022 to 2024)

```python
import pandas as pd

# Load the data
data = pd.read_csv('electric_vehicle_sales_by_state.csv')

# Convert the date column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Filter data for total vehicles for 2022 and 2024
data_2022 = data[(data['date'].dt.year == 2022)]
data_2024 = data[(data['date'].dt.year == 2024)]

# Group by state and sum total vehicles sold for 2022 and 2024
sales_2022 = data_2022.groupby('state')['total_vehicles_sold'].sum().reset_index()
sales_2024 = data_2024.groupby('state')['total_vehicles_sold'].sum().reset_index()

# Merge the sales data for 2022 and 2024
merged_sales = pd.merge(sales_2022, sales_2024, on='state', suffixes=('_2022', '_2024'))

# Calculate CAGR for each state
def calculate_cagr(start, end, periods):
    return (end / start) ** (1 / periods) - 1

# Calculate CAGR for each state
merged_sales['CAGR'] = merged_sales.apply(lambda row: calculate_cagr(row['total_vehicles_sold_2022'], 
                                                                    row['total_vehicles_sold_2024'], 
                                                                    2), axis=1)

# Sort states by CAGR and select the top 10
top_10_states_cagr = merged_sales.sort_values(by='CAGR', ascending=False).head(10)

# Display the results
print("Top 10 States with Highest CAGR in Total Vehicles Sold (2022-2024):")
print(top_10_states_cagr[['state', 'total_vehicles_sold_2022', 'total_vehicles_sold_2024', 'CAGR']])

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='CAGR', y='state', data=top_10_states_cagr, palette='viridis')
plt.title('Top 10 States with Highest CAGR in Total Vehicles Sold (2022-2024)')
plt.xlabel('CAGR')
plt.ylabel('State')
plt.grid(True)
plt.show()
```

#### 8. What are the peak and low season months for EV sales based on the data from 2022 to 2024?
Which means Peak and Low Season Months for EV Sales (2022 to 2024)

```python
# Load the data
data = pd.read_csv('electric_vehicle_sales_by_state.csv')

# Convert the date column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Extract year and month from the date
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month

# Aggregate sales data by month and year
monthly_sales = data.groupby(['year', 'month'])['electric_vehicles_sold'].sum().reset_index()

# Pivot the data to have months as rows and years as columns
monthly_sales_pivot = monthly_sales.pivot(index='month', columns='year', values='electric_vehicles_sold')

# Calculate the average sales for each month across the years
monthly_sales_pivot['average_sales'] = monthly_sales_pivot.mean(axis=1)

# Determine peak and low season months
peak_month = monthly_sales_pivot['average_sales'].idxmax()
low_month = monthly_sales_pivot['average_sales'].idxmin()

# Display the results
print("Peak Season Month (Highest Average Sales):", peak_month)
print("Low Season Month (Lowest Average Sales):", low_month)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales_pivot.index, monthly_sales_pivot['average_sales'], marker='o', linestyle='-', color='b')
plt.title('Average Monthly EV Sales (2022-2024)')
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.xticks(monthly_sales_pivot.index, 
           ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

#### 9. What is the projected number of EV sales (including 2-wheelers and 4- wheelers) for the top 10 states by penetration rate in 2030, based on the compounded annual growth rate (CAGR) from previous years?

Which means Projected Number of EV Sales for Top 10 States by Penetration Rate in 2030

```python
# Load the dataset
data = pd.read_csv('electric_vehicle_sales_by_state.csv')

# Filter data for 2-wheeler and 4-wheeler sales
data['date'] = pd.to_datetime(data['date'])
data['year'] = data['date'].dt.year
data = data[data['year'].isin([2022, 2024])]

# Pivot data to have sales by state and year
pivot_data = data.pivot_table(index='state', columns='year', values='electric_vehicles_sold', aggfunc='sum').reset_index()
pivot_data.columns = ['state', 'sales_2022', 'sales_2024']

# Calculate CAGR
pivot_data['CAGR'] = ((pivot_data['sales_2024'] / pivot_data['sales_2022']) ** (1/2)) - 1

# Identify Top 10 States by Penetration Rate for 2024 (assuming you have penetration rates)
# For demonstration, we sort by sales in 2024
top_10_states = pivot_data.nlargest(10, 'sales_2024')
# Define the number of years from 2024 to 2030
years = 2030 - 2024

# Calculate the projected sales for 2030
top_10_states['projected_sales_2030'] = top_10_states['sales_2024'] * ((1 + top_10_states['CAGR']) ** years)

# Display the results
print("Projected EV Sales for Top 10 States in 2030:")
print(top_10_states[['state', 'projected_sales_2030']])

# Assuming 'top_10_states' DataFrame with 'state' and 'projected_sales_2030'
plt.figure(figsize=(12, 8))
plt.barh(top_10_states['state'], top_10_states['projected_sales_2030'], color='skyblue')
plt.xlabel('Projected Sales in 2030')
plt.ylabel('State')
plt.title('Projected EV Sales for Top 10 States in 2030')
plt.gca().invert_yaxis()  # To display the highest values on top
plt.show()
```

#### 10. Estimate the revenue growth rate of 4-wheeler and 2-wheelers EVs in India for 2022 vs 2024 and 2023 vs 2024, assuming an average unit price. H
Estimate Revenue Growth Rate of EVs (2022 vs. 2024 and 2023 vs. 2024)

```python
# Load the data
data = pd.read_csv('electric_vehicle_sales_by_makers.csv')

# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'], errors='coerce')

# Verify the conversion
print(data['date'].dtype)

# Filter data for the relevant years and vehicle categories
data_2022 = data[data['date'].dt.year == 2022]
data_2023 = data[data['date'].dt.year == 2023]
data_2024 = data[data['date'].dt.year == 2024]

# Separate data for 2-wheelers and 4-wheelers
data_2W_2022 = data_2022[data_2022['vehicle_category'] == '2-Wheelers']
data_2W_2023 = data_2023[data_2023['vehicle_category'] == '2-Wheelers']
data_2W_2024 = data[data['date'].dt.year == 2024][data['vehicle_category'] == '2-Wheelers']

data_4W_2022 = data_2022[data_2022['vehicle_category'] == '4-Wheelers']
data_4W_2023 = data_2023[data_2023['vehicle_category'] == '4-Wheelers']
data_4W_2024 = data_2024[data_2024['vehicle_category'] == '4-Wheelers']

average_price_2W = 50000  # Example price
average_price_4W = 500000  # Example price

# Calculate revenue for each category and year
revenue_2W_2022 = data_2W_2022['electric_vehicles_sold'].sum() * average_price_2W
revenue_2W_2023 = data_2W_2023['electric_vehicles_sold'].sum() * average_price_2W
revenue_2W_2024 = data_2W_2024['electric_vehicles_sold'].sum() * average_price_2W

revenue_4W_2022 = data_4W_2022['electric_vehicles_sold'].sum() * average_price_4W
revenue_4W_2023 = data_4W_2023['electric_vehicles_sold'].sum() * average_price_4W
revenue_4W_2024 = data_4W_2024['electric_vehicles_sold'].sum() * average_price_4W

# Function to calculate growth rate
def calculate_growth_rate(revenue_previous, revenue_current):
    return ((revenue_current - revenue_previous) / revenue_previous) * 100

# Calculate growth rates
growth_rate_2W_2022_2024 = calculate_growth_rate(revenue_2W_2022, revenue_2W_2024)
growth_rate_2W_2023_2024 = calculate_growth_rate(revenue_2W_2023, revenue_2W_2024)

growth_rate_4W_2022_2024 = calculate_growth_rate(revenue_4W_2022, revenue_4W_2024)
growth_rate_4W_2023_2024 = calculate_growth_rate(revenue_4W_2023, revenue_4W_2024)

# Display results
print(f"Growth Rate for 2-Wheelers from 2022 to 2024: {growth_rate_2W_2022_2024}%")
print(f"Growth Rate for 2-Wheelers from 2023 to 2024: {growth_rate_2W_2023_2024}%")

print(f"Growth Rate for 4-Wheelers from 2022 to 2024: {growth_rate_4W_2022_2024}%")
print(f"Growth Rate for 4-Wheelers from 2023 to 2024: {growth_rate_4W_2023_2024}%")


# Data for plotting
years = ['2022', '2023', '2024']
revenue_2W = [revenue_2W_2022, revenue_2W_2023, revenue_2W_2024]
revenue_4W = [revenue_4W_2022, revenue_4W_2023, revenue_4W_2024]

# Plotting
plt.figure(figsize=(12, 6))

plt.plot(years, revenue_2W, marker='o', label='2-Wheelers')
plt.plot(years, revenue_4W, marker='o', label='4-Wheelers')

plt.xlabel('Year')
plt.ylabel('Revenue')
plt.title('Revenue Growth for 2-Wheelers and 4-Wheelers (2022-2024)')
plt.legend()
plt.grid(True)
plt.show()
```
